## goal : We want to scrapp the position ,the address, the name ,the company name of all Mcdonald restaurant in France 

###  method : we will use the mcdonald api 

```python
import requests

# Restaurant URL
base_url = "https://ws.mcdonalds.fr/api/restaurant/"
restaurant_data = []

for restaurant_id in range(0, 2000):
    restaurant_url = f"{base_url}{restaurant_id}"

    try:
        response = requests.get(restaurant_url)
        if response.status_code == 200:
            data = response.json()

            if 'restaurantAddress' in data:
                address = data['restaurantAddress']
                name = data['name']
                coordinates = data['coordinates']
                company_name = data['companyName']

                restaurant_info = {
                    'Address': address,
                    'Name': name,
                    'Coordinates': coordinates,
                    'Company Name': company_name
                }

                restaurant_data.append(restaurant_info)
                print(f"Restaurant Info: {restaurant_info}")
            else:
                print("Address not found in the response.")
        else:
            print(f"Restaurant {restaurant_id} API returned status code: {response.status_code}")
    except Exception as e:
        print(f"An error occurred: {str(e)}")
```

```python
address_list = []
zip_code_list = []
city_list = []
id_list = []
coordinates_list = []
company_name_list = []
name_list = []

for restaurant_entry in restaurant_data:
    address = restaurant_entry['Address'][0]  # Access the first address dictionary
    address_list.append(address['address1'])
    zip_code_list.append(address['zipCode'])
    city_list.append(address['city'])
    id_list.append(address['id'])  # Access the 'id' within the address dictionary
    name_list.append(restaurant_entry['Name'])  # Access 'Name' at the root level
    coordinates_list.append(restaurant_entry['Coordinates'])
    company_name_list.append(restaurant_entry['Company Name'])


print("Addresses:", address_list)
print("Zip Codes:", zip_code_list)
print("Cities:", city_list)
print("IDs:", id_list)
print("Names:", name_list)
```
